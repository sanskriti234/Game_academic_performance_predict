```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv("/content/Gaming_Academic_Performance.csv")
df.head()

```

```python
df.info()
```

```python
df.describe()
```

```python
plt.plot(df.corr(numeric_only=True)['grades'].sort_values(ascending=False),marker='o')
plt.xticks(rotation=90, ha='right')
plt.xlabel("Features")
plt.ylabel("Coefficient Value")
plt.title("Feature Impact on Academic Grades")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['grades'], bins=10)
plt.title('Distribution of Grades')
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True))

plt.title("Correlation Matrix")
plt.show()
```

```python
sns.boxplot(
    x='stress_level',
    y='grades',
    data=df
)
plt.show()
```

```python
sns.pairplot(
    df[['gaming_hours',
        'study_hours',
        'attendance',
        'addiction_score',
        'grades',
        'gender']],hue="gender")
plt.show()
```
